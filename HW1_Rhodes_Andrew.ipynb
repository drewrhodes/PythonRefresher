## ##### Q4. Write a program that prompts the user to enter a loan amount, annual interest rate, and number of years  for a car loan. Then calculate and print the monthly payment amount.

Note:
```
PMT= [PV*i*((1+i)^n)]/((1+i)^n)-1
```
PMT is the monthly payment

PV is the loan amount

i is the interest rate per month in decimal form (interest rate percentage divided by 12)

n is the number of months (term of the loan in months)

```
Loan Amount: 30000.00
Annual Interest Rate: 4.00
Years: 5
Your monthly payment is: $552.50
```

````
def main():
    loan = float(input("Enter a loan amount: "))
    interest = float(input("Enter annual interest rate: "))
    years = int(input("Enter amount of years for loan: "))

    n = years * 12
    r = (interest/100)/12
    d = loan*((r*(1+r)**n)/(((1+r)**n)-1))

    print("Your monthly payment is: $" + str(d))

main()
````

##### Q5. Write a program to that prompts the user to enter a string. Then check whether or not the string is a palindrome.

```
Enter a string: Madam
Madam, is a palindrome!
```

````
def main():
    nameList = []
    reverseNameList = []

    name = input("What is your name? ")

    for a in name:
        nameList.append(a)

    for a in name:
        reverseNameList.insert(0, a)

    nameString = "".join(nameList)
    reverseNameString = "".join(reverseNameList)

    if nameString == reverseNameString:
        print(nameString.capitalize() + ", is a palindrome!")
    else:
        print(nameString.capitalize() + ",  is not a palindrome!")
main()
````

##### Q6. Write a program to compute and print all possible combinations of change for \$1. Denominations to be considered – quarter, dime, nickel, penny.

Hint: Use nested loops (loops within loops for the various denominations of coins)

```
Change for $1:
0 quarters, 0 dimes, 0 nickels, 100 pennies
…
4 quarters, 0 dimes, 0 nickels, 0 pennies
```

````
def main():
    penny = 1
    nickel = 5
    dime = 10
    quarter = 25
    dollar = 100
    counter = 0
    print("Change for $1: \n")
    for q in range(0,5):
        for d in range(0, 11):
            for n in range(0, 21):
                for p in range (0, 101):
                    if q*quarter + d*dime + n*nickel + p*penny == dollar:
                        counter = counter + 1
                        print("Possibility " + str(counter) + " for change for a $1: ")
                        print(str(q) + " quarters, " + str(d) + " dimes, " + str(n) + " nickels, " + str(p) + " pennies.\n")
main()
````

##### Q7. Ask the user to enter two positive integers between 1 and 100. Read those integers. Then output a multiplication table of the first number times the second number. 

Note: You have to ensure the entered numbers are between 1 and 100 only.

```
Please enter an integer: 5
Please enter another integer: 20
5 x 1 = 5
5 x 2 = 10
5 x 3 = 15
…
5 x 20 = 100
```

````
def main():
    firstNum = int(input("Enter a positive integer between 1 and 100: "))
    secondNum = int(input("Enter another positive integer between 1 and 100: "))
    for n in range(1, secondNum+1):
        product = firstNum * n
        print(str(firstNum) + " x " + str(n) + " = " + str(product))
main()
````

##### Q8. Write a program to ask the user to enter a password. Then check to see if it is a valid password based on these requirements -

Must be at least 8 characters long

Must contain both uppercase and lowercase letters

Must contain at least one number between 0-9

Must contain a special character -!,@,#,$

If the password is not valid, ask the user to re-enter. This should continue until the user enters a valid password. After a valid password is entered, print Access Granted!

```
Please enter a password. Follow these requirements - 
a. Must be at least 8 characters long
b. Must contain both uppercase and lowercase letters
c. Must contain at least one number between 0-9
d. Must contain a special character -!,@,#,$
Password: HelloWorld1
Invalid password. Try again!
Password: Hello@World1
Access Granted!
```


````
def main():
    print("Please enter a password. Follow these requirements - \na.	Must be at least 8 characters long\nb.	Must contain both uppercase and lowercase letters\nc.	Must contain at least one number between 0-9\nd.	Must contain a special character: !,@,#,$\n")

    isValid = True
    passwordLength = False
    passwordUpper = False
    passwordLower = False
    passwordNum = False
    passwordSpecial = False

    # While loop to check whether program should keep running and ask the user for another password.
    while isValid == True:
        password = input("Password: ")

        # Check if the password is more than 8 characters, set to true if it passes this condition.
        if len(password) >= 8:
            passwordLength = True

        # loop to go through the entire password and check each character if they have lowercase, uppercase, 0-9 and special characters.
        for char in password:
            if char.islower() == True:
                passwordLower = True
            elif char.isupper() == True:
                passwordUpper = True
            elif char == "0" or char == "1" or char == "2" or char == "3" or char == "4" or char == "5" or char == "6" or char == "7" or char == "8" or char == "9":
                passwordNum = True
            elif char == "!" or char == "@" or char == "#" or char == "$" or char == ",":
                passwordSpecial = True

        # If all of the requirements are met and all the values are True, exit the while loop and print Access Granted!
        if passwordLength == True and passwordUpper == True and passwordLower == True and passwordNum == True and passwordSpecial == True:
            isValid = False

        # If not all the requirements are met, print Invalid Password! Try Again! and loop back through the while loop to prompt the user for another password.
        else:
            print("Invalid password! Please re-enter a valid password!")
    print("Access Granted!")
main()
````
